```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn import linear_model 
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv')
df.head()
```

```python
x = df['year']
```

```python
y = df['per capita income (US$)']
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income (USD)')
plt.scatter(x,y,color='red',s=20, label='PCI(USD) By Year')
plt.legend()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],y)
```

```python
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income (USD)')
plt.scatter(x,y,color='red',s=20, label='PCI(USD) By Year')
plt.plot(x,reg.predict(df[['year']]),color='blue',label='Predicted PCI(USD)')
plt.legend()
```
