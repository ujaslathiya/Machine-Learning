```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/19_Bagging/Exercise/heart.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
# df['ChestPainType'] = le.fit_transform(df['ChestPainType'])
# df
```

```python
dummies = pd.get_dummies(df,drop_first=True)
```

```python
dummies.head()
```

```python
dummies.drop(['Age','RestingBP','Cholesterol','FastingBS','MaxHR','Oldpeak','HeartDisease'],axis='columns',inplace=True)
```

```python
dummies.head()
```

```python
df = pd.concat([df,dummies],axis='columns')
```

```python
pd.set_option('display.max_columns',26)
```

```python
df
```

```python
df.drop(['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope'],axis='columns',inplace=True)
```

```python
X = df.drop(['HeartDisease'],axis='columns')
y = df.HeartDisease
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,stratify=y,random_state=10)
```

```python
from sklearn.svm import SVC
svc = SVC()

svc.fit(X_train,y_train)
```

```python
svc.score(X_test,y_test)
```

```python
from sklearn.ensemble import BaggingClassifier
beg_clf = BaggingClassifier(
    estimator=SVC(),
    n_estimators=100,
    random_state=0,
    max_samples=0.8
)

beg_clf.fit(X_train,y_train)
```

```python
beg_clf.score(X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
dt.score(X_test,y_test)
```

```python
from sklearn.ensemble import BaggingClassifier
beg_clf = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0    
)

beg_clf.fit(X_train,y_train)
```

```python
beg_clf.oob_score_
```

```python
beg_clf.score(X_test,y_test)
```
